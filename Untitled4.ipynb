```python
import random

def make_portfolios(df):
    df = df.append(df, ignore_index=True)
    with_claims = []
    no_claims = []
    for i, r in enumerate(df['Loss_Amount']):
        if r > 0.0:
            with_claims.append(i)
        else:
            no_claims.append(i)
            
    random.shuffle(with_claims)
    random.shuffle(no_claims)
    portfolios = []
    i = 0
    while True:        
        size = np.random.choice(range(1, 5), p=[0.2, 0.3, 0.3, 0.2]) * 500
        loss_percent = np.random.choice(range(0, 5), p=[0.2]*5) * 5
        if loss_percent == 0:
            loss_percent = 2
                
        portfolio = []
        num_losses = int(loss_percent*(size/100))
#         print(i,'-',size, loss_percent, num_losses)
        i += 1
        for i in range(num_losses):
            try:
                portfolio.append(df.iloc[with_claims.pop()])
            except Exception as e:
#                 print('breaking,', e)
                break

        for i in range(size-num_losses):
            try:
                portfolio.append(df.iloc[no_claims.pop()])
            except Exception as e:
#                 print('breaking,', e)
                break        

        if len(with_claims) == 0 or len(no_claims) == 0:
#             if len(with_claims) != 0:
#                 while(len(with_claims) > 0):
#                     portfolio.append(df.iloc[with_claims.pop()])
#                 portfolios.append(pd.DataFrame(portfolio, columns=df.columns))
#             if len(no_claims) != 0:
#                 while(len(no_claims) > 0):
#                     portfolio.append(df.iloc[no_claims.pop()])
#                 portfolios.append(pd.DataFrame(portfolio, columns=df.columns))                
            break
            
        random.shuffle(portfolio)
        portfolios.append(pd.DataFrame(portfolio, columns=df.columns))
    return portfolios
```

```python
def median_summarizer(df_t):
    values = df_t.median()
    n_df = pd.DataFrame([], columns=df_t.columns)
    return n_df.append(values, ignore_index=True)

def mean_summarizer(df_t):
    values = df_t.mean()
    n_df = pd.DataFrame([], columns=df_t.columns)
    return n_df.append(values, ignore_index=True)

def max_summarizer(df_t):
    values = df_t.max()
    n_df = pd.DataFrame([], columns=df_t.columns)
    return n_df.append(values, ignore_index=True)

def min_summarizer(df_t):
    values = df_t.min()
    n_df = pd.DataFrame([], columns=df_t.columns)
    return n_df.append(values, ignore_index=True)

def min_max_summarizer(df_t):
    values_min = df_t.min()
    values_max = df_t.max()
    values = (values_max + values_min)/2
    n_df = pd.DataFrame([])
    return n_df.append(values, ignore_index=True)

def std_dev_summarizer(df_t):
    values = df_t.std()
    n_df = pd.DataFrame([])
    return n_df.append(values, ignore_index=True)

def num_median_summarizer(df_t):
    df_med = median_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_med = df_med.values[0]
    values = [0] * len(df_med)
    for r in df_t.values:
        for i in range(len(r[:-2])):
            if r[i] > df_med[i]:
                values[i] += 1
    values = [v/len(df_t)*10 for v in values]
    #     print(len(values))
    n_df = pd.DataFrame([])
    return n_df.append([values], ignore_index=True)


def num_mean_std_dev_summarizer(df_t):
    df_std = std_dev_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_mean = mean_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_std = df_std.values[0]
    df_mean = df_mean.values[0]
    values = [0] * len(df_std)
    for r in df_t.values:
        for i in range(len(r[:-2])):
            if r[i] > df_mean[i] + df_std[i]:
                values[i] += 1
    values = [v/len(df_t)*10 for v in values]
#     print(len(values))
    n_df = pd.DataFrame([])
    return n_df.append([values], ignore_index=True)

def num_median_std_summarizer(df_t):
    df_med = median_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_std = num_mean_std_dev_summarizer(df_t)
    n_df = df_std
    n_df[[len(n_df.columns) + i for i in range(30)]] = df_med
    return n_df

def all_summarizer(df_t):
    df_std = std_dev_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_max = max_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_min = min_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_med = median_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    df_men = mean_summarizer(df_t.drop(['Loss_Amount', 'Annual_Premium'], axis=1))
    
    n_df = df_std
    n_df[[len(n_df.columns) + i for i in range(40)]] = df_max
    n_df[[len(n_df.columns) + i for i in range(40)]] = df_min
    n_df[[len(n_df.columns) + i for i in range(40)]] = df_med
    n_df[[len(n_df.columns) + i for i in range(40)]] = df_men
    return n_df
    

def summarize_portfolios(portfolios, summarizer=median_summarizer, convert_to_lr=False):
    summarized_portfolios = []
    flag = True
    for p in portfolios:
        portfolio = summarizer(p)
        if convert_to_lr:
            portfolio['lr'] = np.log(p['Loss_Amount'].sum()/p['Annual_Premium'].sum())
#             print(len(portfolio.columns))
            portfolio = portfolio.drop(['Loss_Amount', 'Annual_Premium'], axis=1)
        if flag:
            cols = portfolio.columns
            flag = False
        summarized_portfolios.append(portfolio)
        
    if cols is None:
        return None
    n_df = pd.DataFrame([], columns=cols)
    for p in summarized_portfolios:
        n_df = n_df.append(p, ignore_index=True)
#     print(n_df)

#     pca_p = PCA().fit(n_df.drop('lr', axis=1))
#     principal_cols = pca_p.transform(n_df.drop('lr', axis=1))
#     n_df_t = pd.DataFrame(data = principal_cols)

    n_df_arr = StandardScaler().fit_transform(n_df.drop('lr', axis=1))
    n_df_t = pd.DataFrame(n_df_arr)
    n_df_t['lr'] = n_df['lr']
    
    return n_df_t
```
