```python
X = pd.read_csv('train_half_median_40.csv')
T = pd.read_csv('test_half_median_40.csv')
Y = pd.read_csv('train_lr_half.csv', header=None)
true_pred = pd.read_csv('test_lr_half.csv', header=None)
```

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(max_iter=len(X), hidden_layer_sizes=[100,], activation='tanh', 
                   learning_rate='adaptive', learning_rate_init=0.1, shuffle=True)
```

```python
mlp.fit(X, Y)
```

```python
mlp.score(T, true_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(true_pred, mlp.predict(T))
```

```python
T_n = T[T.columns[:]]
X_n = X[X.columns[:]]
```

```python
T_n
```

```python
X_n
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2)
X_np = poly.fit_transform(X_n)
X_np = pd.DataFrame(X_np)
T_np = poly.fit_transform(T_n)
X_np = pd.DataFrame(X_np)
```

```python
X_np
```

```python
from sklearn.linear_model import LinearRegression
reg_l = LinearRegression().fit(X_np, Y)
reg_l.score(T_np, true_pred)
```

```python
mean_absolute_error(true_pred, reg_l.predict(T_np))
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(X_n, Y)
neigh.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, neigh.predict(T_n))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(0.5)
ridge.fit(X_n, Y)
ridge.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, ridge.predict(T_n))
```

```python
from sklearn import linear_model
breg = linear_model.BayesianRidge()
breg.fit(X_n, Y)
breg.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, breg.predict(T_n))
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_r = DecisionTreeRegressor(max_depth=1, random_state=0)
tree_r.fit(X_n, Y)
tree_r.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, tree_r.predict(T_n))
```

```python
from sklearn.svm import SVR
svm_r = SVR(gamma='scale', C=1.0, epsilon=0.2)
svm_r.fit(X_n, Y)
svm_r.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, svm_r.predict(T_n))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
rf.fit(X_n, Y)
rf.score(T_n, true_pred)
```

```python
mean_absolute_error(true_pred, rf.predict(T_n))
```
